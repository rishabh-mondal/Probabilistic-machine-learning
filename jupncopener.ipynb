```python
from matplotlib import pyplot as plt
import numpy as np
time_step_index = 0

# Extract TSURF_K data for the specified time step
tsurf_data = datax['TSURF_K'].isel(TSTEP=time_step_index,LAY=0)

# Plot TSURF_K data for all rows and columns
plt.figure(figsize=(10, 6))
plt.contourf(tsurf_data.COL, tsurf_data.ROW, tsurf_data.values, cmap='coolwarm')
plt.colorbar(label='Surface Temperature (K)')
plt.xlabel('COL')
plt.ylabel('ROW')
plt.title(f'TSURF_K Data for Time Step {time_step_index}')
plt.show()
```

```python
import matplotlib.animation as animation
import imageio
num_time_steps = len(datax.TSTEP)
print(num_time_steps)
```

```python
time_step_index = 0
def save_tsurf_plot(time_step_index):
    tsurf_data = datax['TSURF_K'].isel(TSTEP=time_step_index,LAY=0)
    # Plot TSURF_K data for all rows and columns
    fig, ax = plt.subplots(figsize=(10, 6))
    plt.contourf(tsurf_data.COL, tsurf_data.ROW, tsurf_data.values, cmap='coolwarm')
    plt.colorbar(label='Surface Temperature (K)')
    plt.xlabel('COL')
    plt.ylabel('ROW')
    plt.title(f'TSURF_K Data for Time Step {time_step_index}')
    plt.savefig(f'tsurf20230812/tsurf_plot_{time_step_index}.png')  # Save the plot as an image
    plt.close()
for i in range(num_time_steps):
    save_tsurf_plot(i)

images = []
for i in range(num_time_steps):
    image_path = f'tsurf20230812/tsurf_plot_{i}.png'
    images.append(imageio.imread(image_path))

# Save the images as a GIF animation
imageio.mimsave('tsurf_animation.gif', images, fps=5)
import os
for i in range(num_time_steps):
    image_path = f'tsurf20230812/tsurf_plot_{i}.png'
    os.remove(image_path)
```
