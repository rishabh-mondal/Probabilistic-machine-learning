```python
def get_latitudes():
    lat_start= 76.8499984741211
    lat_step=0.009999999776482582

    latitudes=[]

    for i in range(80):
        latitudes.append(lat_start+i*lat_step)
    
    latitudes.reverse()

    return latitudes

```

```python
def get_longitudes():
    long_start= 28.200000762939453
    long_step=0.009999999776482582

    longitudes=[]

    for i in range(80):
        longitudes.append(long_start+i*long_step)
    
    # longitudes.reverse()

    return longitudes
```

```python
def create_p10_25_plot(day,var_name):
    p10_hour=data[var_name][data['TSTEP']==day] 
    p10_hour=p10_hour[0,:,:]
    p10_hour=p10_hour[0,:,:]
    plt.imshow(p10_hour)
    plt.title(f'{var_name} at hour '+str(day))
    plt.colorbar()

    # only show every latitude and longitude of end points

    latitudes=get_latitudes()
    longitudes=get_longitudes()

    # round to 2 decimal places

    top=latitudes[0]
    top=round(top,2)
    bottom=latitudes[-1]
    bottom=round(bottom,2)
    left=longitudes[0]
    left=round(left,2)
    right=longitudes[-1]
    right=round(right,2)

    plt.xticks([0,79],[left,right])
    plt.yticks([0,79],[top,bottom])

    # plt.savefig(f'plots/120/{var_name}_{day}.png')

    # plt.close()
```

```python
create_p10_25_plot(2,'P10')
```
